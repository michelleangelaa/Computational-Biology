```python
import Bio
from Bio import SeqIO
from Bio.SeqUtils import Seq, GC, gc_fraction, MeltingTemp, molecular_weight
from Bio.Data import CodonTable
```

```python
from Bio.SeqUtils import MeltingTemp as MT
```

```python
simple_sequence = Seq("ATGATCTCGTAA")
```

```python
melting_wallace = MT.Tm_Wallace(simple_sequence)
melting_NN = MT.Tm_NN(simple_sequence)
melting_GC = MT.Tm_GC(simple_sequence)
print(f"Output: Wallace: {melting_wallace}, NN: {melting_NN}, GC: {melting_GC}")
```

# Protein Synthesis

```python
# transcription -> transcribe()
mRNA = simple_sequence.transcribe()
mRNA
```

```python
# translation -> translate()
protein = mRNA.translate()
protein
```
