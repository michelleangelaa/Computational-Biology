# Manipulasi Sequence

```python
# genbank
gb = SeqIO.read("files\sequence.gb", "genbank")
seq_gb = gb.seq
print(seq_gb)
```

```python
print(f'Seq_gb: ', seq_gb)
```

```python
# [start : end : jump]
print(seq_gb[::2])

# reverse
print(seq_gb[::-1])
```

```python
# buat sequence jadi 2 bagian (slicing)
seq1 = seq_gb[:round(len(seq_gb)/2)]
seq2 = seq_gb[round(len(seq_gb)/2):]

print(f'Seq1 (setengahnya seq_gb dari awal sampai tengah) : ', seq1)
print(f'Seq2 (Setengahnya seq_gb dari tengah sampai akhir) : ', seq2)
```

```python
# misal disuruh reverse seq2 trus join sama seq1
new_seq = seq2[::-1]+seq1
print('hasil merge: ', new_seq)
```

```python
# misal disuru ganti semua GC jadi AT (replace)
seq2_replace = seq2.replace("GC", "AT")
print(f"Seq2_replace: ", seq2_replace)
```
